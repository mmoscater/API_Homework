```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from urllib.error import HTTPError
from pprint import pprint
# openweathermap lib
import openweathermapy as ow


# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# use openweathermapy instead of using url
# create sets of 50 
n=50

# break list into sublists
sets = [cities[i * n:(i + 1) * n] for i in range((len(cities) + n - 1) // n )]  
#print (len(sets)) 

# set settings
settings = {"units":"imperial","appid":api_keys.api_key}

# set weather data list
weather_data = []


print(f'Beginning Data Retrieval:\n---------------------------')
for i,cg in enumerate(sets,start=0):
    set_x = cg
    counter = 1
    #print(i)
    for city in cg:
        #print(city)
        print(f'Processing Record {counter} of Set {i+1} | {city}')
        counter = counter +1
        try:
            weather_data.append([ow.get_current(city,**settings)])
        except(HTTPError):
            print(f'City not found. Skipping...')
    #takes ~10 seconds per group (call limit is 60 pm)        
    time.sleep(50) 
print(f'---------------------------\nData Retrieval Complete!\n---------------------------')

            
        
```

```python
# check to make sure there are 500 records
print(len(weather_data))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# look at structure to set up summary list
pprint(weather_data[0])
```

```python
# set up summary
summary = ["name","clouds.all","sys.country","dt","main.humidity","coord.lat"
          ,"coord.lon","main.temp_max","wind.speed"]
column_head = ['City','Cloudiness','Country','Date','Humidity'
          ,'Lat','Lng','Max Temp','Wind Speed']
# loop for specific elements
w_data = [resp[0](*summary) for resp in weather_data]
#print(w_data)
# w_data = [d[0]['name'] for d in weather_data]
# print(w_data[1])
# create data frame
weather_df = pd.DataFrame(w_data,columns=column_head)
weather_df.to_csv("Random_City_Weather.csv")
weather_df.head()
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
# set x and y for scatter plot
lat = weather_df["Lat"]
temp = weather_df["Max Temp"]
xl = 'Latitude'
# set plot
plt.scatter(lat,temp,facecolor='orange',edgecolor='black',alpha=0.5)
plt.grid()
plt.title(f'City Latitude vs Max Temperature (10/27/18)')
plt.xlabel(xl)
plt.xlim(-90,90)
plt.ylabel('Max Temperature (F)')
plt.savefig('LatVsTemp.png')
plt.show()
```

#### Latitude vs. Humidity Plot

```python
# set humidity
hum = weather_df["Humidity"]

# plot
plt.scatter(lat,hum,facecolor='coral',edgecolor='black',alpha=0.5)
plt.grid()
plt.title('City Latitude vs Humidity (10/27/18)')
plt.xlabel(xl)
plt.ylabel('Humidity (%)')
plt.xlim(-90,90)
plt.savefig('LatVsHumid.png')
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
# set cloudiness
cloud = weather_df['Cloudiness']

# plot 
plt.scatter(lat,cloud,facecolor='skyblue',edgecolor='black',alpha=0.5)
plt.grid()
plt.xlim(-90,90)
plt.xlabel(xl)
plt.ylabel('Cloudiness (%)')
plt.title('City Latitude vs. Cloudiness (10/27/18)')
plt.savefig('LatVsCloud.png')
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# set windspeed
wind = weather_df['Wind Speed']

# plot 
plt.scatter(lat,wind,facecolor='green',edgecolor='black',alpha=0.5)
plt.grid()
plt.title('City Latitude vs Wind Speed (10/27/18)')
plt.xlim(-90,90)
plt.xlabel(xl)
plt.ylabel("Wind Speed (mph)")
plt.savefig('LatVsWind.png')
plt.show()
```
